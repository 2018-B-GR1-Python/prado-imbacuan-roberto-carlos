## 2.2 Obtenemos la distribución de tempos de la lista
Con este paso deseamos saber cuales son los tempos mas comunes en la lista de las mejores canciones.

```python
data = top_songs['tempo'].values
print(data.shape)
(n, bin_edges) = np.histogram(data, bins=50)
plt.plot(bin_edges[1:],n)
plt.show()
```

```python
freq1 = (top_songs.loc[:,"tempo"] > 95) & (top_songs.loc[:,"tempo"] < 105)
freq2 = (top_songs.loc[:,"tempo"] > 125) & (top_songs.loc[:,"tempo"] < 135)
```

# Hipotesis 3
*Las canciones mas populares se originan en un lugar en particular*

## 3.1 Obtenemos la lista con un puntaje mayor a 0.8

```python
indices_canciones_mas_populares = music_dataset.loc[:,'song_hotttnesss'] > 0.8
canciones_rank = music_dataset[indices_canciones_mas_populares]
canciones_rank
```

## 3.1 Obtenemos la lista con las localizaciones

```python
canciones_ubicadas_duplicadas = canciones_rank.loc[:,('location','latitude','longitude')]
canciones_ubicadas_duplicadas
```

## 3.3 Obtenemos las frecuencias de cada ubicación (Agrupación)
Existen varias coordenadas duplicadas, realiazamos un conteo para cada coordenada única. 

```python
df = canciones_ubicadas_duplicadas.groupby(['latitude', 'longitude']).size().reset_index(name='Freq')
print (df)
```

## 3.4 Ordenamos la lista en forma descendente

```python
df_sort = df.sort_values('Freq', ascending=False)
df_sort
```

## 3.6 Agrupamos por region

```python
lat_freq = df_sort.groupby(pd.cut(df_sort["latitude"], np.arange(15, 60.0+5, 5))).count()
lon_freq = df_sort.groupby(pd.cut(df_sort["longitude"], np.arange(-130, 10.0+10, 10))).count()
print(lat_freq.loc[:,['latitude','Freq']])
print(lon_freq.loc[:,['longitude','Freq']])
```

## 3.7 Teoría
Las canciones mas populares tienden a ser mas frecuentes en un solo lugar, sin embargo esta coordenada (37.157357, -63.933358) corresponde a una posición no disponible, para los demas valores la diferencia es minima, siendo el valor de frecuencia mas alto 3 y el mas bajo 1 sobre un total de 104 canciones, por lo tanto la ubicación de origen de la canción de influye en su popularidad

## 4.1 División en 4 regiones



De acuerdo al mapa, dividiremos al mapa en cuatro partes definidas por la linea ecuatorial y el meridiano de  greenwich

## 4.2 Obtenemos las canciones de las regiones: Norte America, Sudamerica, Europa Asia, Africa y Oceania

```python
SUD_AMERICA_df = music_dataset.query('latitude < 0 & longitude < 0')
SUD_AMERICA_df
```

```python
EUROPA_ASIA_df = music_dataset.query('latitude > 0 & longitude > 0')
EUROPA_ASIA_df
```

```python
AFRICA_OCEAN_df = music_dataset.query('latitude < 0 & longitude > 0')
AFRICA_OCEAN_df
```

## 4.3 Promedio del campo ruido en cada dataframe

```python
cols = ['Ruido']
indexes = ['Norte America', 'Sur America', 'Europa-Asia', 'Africa-Oceania']
data_loud = [
    NORTE_AMERICA_df['loudness'].mean(), 
    SUD_AMERICA_df['loudness'].mean(),
    EUROPA_ASIA_df['loudness'].mean(), 
    AFRICA_OCEAN_df['loudness'].mean()]
regiones = pd.DataFrame(data_loud, index=indexes, columns=cols)
regiones
```
